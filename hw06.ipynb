### Write your code below

```python
# YOUR CODE HERE
mauna_loa_co2 = np.genfromtxt('co2_mm_mlo.txt')

x = mauna_loa_co2[:,2]
y = mauna_loa_co2[:,4]
plt.plot(x, y, label = "line")
plt.tight_layout()

p1 = np.polyfit(x, y, 1)
y1 = np.polyval(p1, x)
plt.plot(x, y1, label = "1st degree")

p3 = np.polyfit(x, y, 3)
y3 = np.polyval(p3, x)
plt.plot(x, y3, label = "3rd degree")

p5 = np.polyfit(x, y, 5)
y5 = np.polyval(p5, x)
plt.plot(x, y5, label = "5th degree")

plt.title('1st, 3rd, 5th degree Polynomial Fits')
plt.xlabel('Date')
plt.ylabel('CO2 value')
plt.grid(True, which='both')

plt.legend(loc = 4)
plt.show()
plt.close()

# --------------------

s = np.linspace(x.min(), 2150, num=len(x))

p1 = np.polyfit(x, y, 1)
y1 = np.polyval(p1, s)
plt.plot(s, y1, label = "1st degree", color = "orange")

p3 = np.polyfit(x, y, 3)
y3 = np.polyval(p3, s)
plt.plot(s, y3, label = "3rd degree", color = "green")

p5 = np.polyfit(x, y, 5)
y5 = np.polyval(p5, s)
plt.plot(s, y5, label = "5th degree", color = "red")
plt.tight_layout()

plt.title('Polynomial Fits predicting to 2150')
plt.xlabel('Date')
plt.ylabel('CO2 value')
plt.grid(True, which='both')

plt.legend(loc = 4)
plt.show()
plt.close()

print("Function values for each of the three polynomials at 2150:")
print(f"1st degree: {y1[-1]}")
print(f"3rd degree: {y3[-1]}")
print(f"5th degree: {y5[-1]}")

# print(mauna_loa_co2)
# raise NotImplementedError()
```
